### Provider

Import charges concatenate the dataframes and sum th charges per provider no, then merge with provider dataFrame

Import region to merge and add the region column to provider

```python
def get_concat_charges():
    charges = pd.read_csv("Dataset/charge1.csv", sep=',')
    for i in range(2, 17):
        charges_2 = pd.read_csv(f"Dataset/charge{i}.csv", sep=',')
        charges = pd.concat([charges, charges_2], ignore_index=True, sort=False)
    return charges
```

```python
charges = get_concat_charges()
```

```python
charges
```

```python
charge_amount = charges.groupby("provider_no")["charge_amt"].sum().reset_index()
charge_amount
```

```python
provider_def = pd.read_csv("Dataset/provider.csv", sep=",")
provider_def
```

```python
provider_def_2 = provider_def.merge(charge_amount, how="left", on="provider_no")
provider_def_2
```

Import region

```python
region_df = pd.read_csv("Dataset/region.csv", sep=',')
region_df
```

```python
provider_final_df = provider_def_2.merge(region_df[['region_no', 'region_name']], how="left", on="region_no").drop(columns=["region_no"]).fillna(0).rename(columns={"charge_amt":"charge_amt_sum", "region_name": "region"})
```
